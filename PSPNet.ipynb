```python
# Import necessary libraries
import os
import time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
import requests
from io import BytesIO
import tensorflow as tf
from tensorflow.keras import layers, models, applications, losses, optimizers

# Check for GPU availability
gpus = tf.config.list_physical_devices('GPU')
if gpus:
   print(f"GPUs found: {len(gpus)}")
   for gpu in gpus:
       print("  ", gpu)
else:
   print("No GPU found. Please enable GPU mode for training.")

# 1. Define Constants and Hyperparameters
IMG_HEIGHT = 120
IMG_WIDTH = 120
BATCH_SIZE = 16
EPOCHS = 5              # Adjust number of epochs as needed
DATA_LIMIT = 20           # Limit the number of images for demo
NUM_CLASSES = 2            # For binary segmentation: background (0) and card (1)

# 2. Load the CSV file containing Pokémon card data
# NOTE: You'll need to upload 'pokemon-cards.csv' to your Colab environment.
csv_path = "pokemon-cards.csv"  # adjust path if needed
try:
    df = pd.read_csv(csv_path)
    print("CSV file read successfully. Number of records:", len(df))
    # Use a subset to save computation (for demonstration)
    df = df.head(DATA_LIMIT)
    print("Using a subset of the dataset:", df.shape)
except FileNotFoundError:
    print(f"Error: The file '{csv_path}' was not found. Please upload the CSV to your Colab environment.")
    df = pd.DataFrame() # Create an empty DataFrame to prevent errors later

# 3. Define Image Download, Preprocessing, and Synthetic Mask Generation
def download_and_preprocess_image(url):
   """
   Downloads an image from a URL, decodes it, resizes it to (IMG_HEIGHT, IMG_WIDTH),
   normalizes it, and returns the image.
   """
   try:
       response = requests.get(url, timeout=10)
       response.raise_for_status()
       img_array = np.frombuffer(response.content, np.uint8)
       img = cv2.imdecode(img_array, cv2.IMREAD_COLOR)
       if img is None:
           raise ValueError("Downloaded image is None")
       # Convert from BGR to RGB and resize
       img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
       img = cv2.resize(img, (IMG_WIDTH, IMG_HEIGHT))
       img = img.astype('float32') / 255.0
       return img
   except Exception as e:
       print(f"Error processing {url}: {e}")
       return None

def generate_dummy_mask(img):
   """
   Generates a synthetic binary segmentation mask from an image.
   In this demo, we convert the image to grayscale and threshold it.
   Pixels above a fixed threshold are considered foreground.
   The output mask is of shape (IMG_HEIGHT, IMG_WIDTH) with integer labels {0,1}.
   """
   # Convert image to grayscale
   gray = cv2.cvtColor((img*255).astype('uint8'), cv2.COLOR_RGB2GRAY)
   # Apply a threshold to create a binary mask
   _, mask = cv2.threshold(gray, 120, 255, cv2.THRESH_BINARY)
   # Convert mask to 0/1
   mask = (mask > 0).astype('int32')
   return mask

def data_generator(df):
   """
   For each row in the CSV, downloads the image and generates a dummy mask.
   Yields a tuple: (image, mask) where:
     - image shape: (IMG_HEIGHT, IMG_WIDTH, 3)
     - mask shape: (IMG_HEIGHT, IMG_WIDTH) with values 0 or 1.
   """
   if df.empty:
       return # Exit if DataFrame is empty
   for idx, row in df.iterrows():
       url = row["image_url"]
       img = download_and_preprocess_image(url)
       if img is not None:
           mask = generate_dummy_mask(img)
           yield img, mask

# Create a tf.data.Dataset for segmentation:
dataset = tf.data.Dataset.from_generator(
   lambda: data_generator(df),
   output_types=(tf.float32, tf.int32),
   output_shapes=((IMG_HEIGHT, IMG_WIDTH, 3), (IMG_HEIGHT, IMG_WIDTH))
)

dataset = dataset.shuffle(buffer_size=DATA_LIMIT).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)

# 4. Build the PSPNet Model using a ResNet50 Backbone and a Pyramid Pooling Module
def pyramid_pooling_module(input_tensor, pool_sizes):
   """
   Creates a pyramid pooling module.
   - input_tensor: feature map from backbone
   - pool_sizes: list of grid sizes (e.g., [1, 2, 3, 6])
   Returns the concatenated result of upsampled pooled features and the original input.
   """
   input_shape = tf.keras.backend.int_shape(input_tensor)
   concat_list = [input_tensor]
   h, w = input_shape[1], input_shape[2]
   for pool_size in pool_sizes:
       # Average pooling: kernel size and strides to obtain (pool_size x pool_size)
       x = layers.AveragePooling2D(pool_size=(h//pool_size, w//pool_size),
                                   strides=(h//pool_size, w//pool_size),
                                   padding='same')(input_tensor)
       x = layers.Conv2D(512, kernel_size=1, padding='same', use_bias=False)(x)
       x = layers.BatchNormalization()(x)
       x = layers.Activation('relu')(x)
       # Upsample to original spatial dimensions using bilinear interpolation
       x = layers.UpSampling2D(size=(h//pool_size, w//pool_size), interpolation='bilinear')(x)
       concat_list.append(x)
   return layers.Concatenate()(concat_list)

def build_pspnet(input_shape=(IMG_HEIGHT, IMG_WIDTH, 3), num_classes=NUM_CLASSES):
   """
   Builds a simplified PSPNet model:
     1. Uses a ResNet50 backbone (pre-trained on ImageNet, with output stride reduced by using dilated convolutions if desired).
     2. Applies a pyramid pooling module on the extracted features.
     3. Outputs a segmentation map with dimensions matching the input.
   """
   img_input = layers.Input(shape=input_shape)
   # Use ResNet50 as backbone; freeze or fine-tune as needed.
   base_model = applications.ResNet50(include_top=False, weights='imagenet', input_tensor=img_input)
   # Choose a feature layer (e.g., 'conv4_block6_out' typically gives an output of ~8x8 if input is 120x120)
   backbone_output = base_model.get_layer("conv4_block6_out").output
   # Pyramid Pooling Module with grids: 1x1, 2x2, 3x3, 6x6
   ppm = pyramid_pooling_module(backbone_output, pool_sizes=[1, 2, 3, 6])
   # Fuse the pooled features
   x = layers.Conv2D(512, kernel_size=3, padding='same', use_bias=False)(ppm)
   x = layers.BatchNormalization()(x)
   x = layers.Activation('relu')(x)
   x = layers.Dropout(0.1)(x)
   # Final classification: map features to number of classes
   x = layers.Conv2D(num_classes, kernel_size=1)(x)
   # **Modified UpSampling:** Upsample the logits to input image size.
   # Since the backbone output is 8x8, scale factor = 120 / 8 = 15.
   x = layers.UpSampling2D(size=(15, 15), interpolation='bilinear')(x)
   output = layers.Activation('softmax')(x)
   model = models.Model(inputs=img_input, outputs=output)
   return model

# Build the model
if not df.empty:
    model = build_pspnet()
    model.summary()

# 5. Define Loss, Metrics, and Optimizer
# We use Sparse Categorical Crossentropy (ground truth masks are integer labels)
loss_fn = losses.SparseCategoricalCrossentropy()
optimizer = optimizers.Adam(learning_rate=1e-4)

# Define metrics: pixel accuracy and mean Intersection over Union (IoU)
pixel_acc = tf.keras.metrics.SparseCategoricalAccuracy()
mean_iou = tf.keras.metrics.MeanIoU(num_classes=NUM_CLASSES)

# 6. Training Step
@tf.function
def train_step(images, masks):
   """
   Runs one training step.
   - images: input card images, shape (BATCH_SIZE, IMG_HEIGHT, IMG_WIDTH, 3)
   - masks: corresponding ground truth segmentation masks, shape (BATCH_SIZE, IMG_HEIGHT, IMG_WIDTH)
   """
   with tf.GradientTape() as tape:
       predictions = model(images, training=True)
       loss = loss_fn(masks, predictions)
   gradients = tape.gradient(loss, model.trainable_variables)
   optimizer.apply_gradients(zip(gradients, model.trainable_variables))
   pixel_acc.update_state(masks, predictions)
   mean_iou.update_state(masks, tf.argmax(predictions, axis=-1, output_type=tf.int32))
   return loss

# 7. Training Loop
def train_model(dataset, epochs):
    if not dataset:
        print("Dataset is empty. Skipping training.")
        return
    for epoch in range(epochs):
        start_time = time.time()
        pixel_acc.reset_state()
        mean_iou.reset_state()
        total_loss = 0.0
        steps = 0
        for images, masks in dataset:
            loss = train_step(images, masks)
            total_loss += loss
            steps += 1
        if steps > 0:
            avg_loss = total_loss / steps
            print(f"Epoch {epoch+1}/{epochs}: Loss = {avg_loss:.4f}, Pixel Acc = {pixel_acc.result():.4f}, Mean IoU = {mean_iou.result():.4f}, Time = {time.time()-start_time:.2f} sec")
            # Optionally, every 10 epochs, visualize predictions on a batch
            # if (epoch + 1) % 10 == 0:
            visualize_predictions(dataset, model)

def visualize_predictions(dataset, model, num_examples=3):
   # Get one batch from the dataset
   for images, masks in dataset.take(1):
       preds = model(images, training=False)
       preds = tf.argmax(preds, axis=-1)
       preds = preds.numpy()
       images = images.numpy()
       masks = masks.numpy()
       break
   plt.figure(figsize=(15, 6))
   for i in range(num_examples):
       # Show input image
       plt.subplot(3, num_examples, i+1)
       plt.imshow(images[i])
       plt.title("Input Image")
       plt.axis("off")
       # Show ground truth mask
       plt.subplot(3, num_examples, i+1+num_examples)
       plt.imshow(masks[i], cmap='gray')
       plt.title("Ground Truth")
       plt.axis("off")
       # Show predicted segmentation
       plt.subplot(3, num_examples, i+1+2*num_examples)
       plt.imshow(preds[i], cmap='gray')
       plt.title("Prediction")
       plt.axis("off")
   plt.suptitle("Segmentation Results")
   plt.show()

# 8. Run the Training Process
if not df.empty:
    train_model(dataset, EPOCHS)
```
